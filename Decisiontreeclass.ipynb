```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('dataset.csv')
df.head()
```

```python
df['Gender']=df['Gender'].map({'Male':1,'Female':0})
df.head()
```

```python
#selection of features
x = df[['Weight','Gender']] # Changed tuple to a list
y = df['Likeness']
```

```python
y.head()
```

```python
#Now Ml
from sklearn.tree import DecisionTreeClassifier

lr=DecisionTreeClassifier()
lr.fit(x,y)
```

```python
lr.predict([[14,1]])
```

```python
#measure accuracy
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)

lr=DecisionTreeClassifier()
lr.fit(x_train,y_train)
predicted=lr.predict(x_test)
predicted

score=accuracy_score(y_test,predicted)
score
```

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
import joblib

lr=DecisionTreeClassifier().fit(x,y)
joblib.dump(lr,'foodie.joblib')
```

```python
#now how to run this
model = joblib.load('foodie.joblib')
```

```python
from sklearn import tree
model = DecisionTreeClassifier().fit(x,y)

#graph evaluation
tree.export_graphviz(model,
                     out_file='foodie.dot',
                     feature_names=['age','gender'],
                     class_names=sorted(y.unique()),
                     label='all',
                     rounded=True,
                     filled=True)
```

```python
import graphviz

# Read the .dot file (replace 'foodie.dot' with your actual file name)
with open('foodie.dot', 'r') as file:
    dot_content = file.read()

# Visualize the g/raph
graph = graphviz.Source(dot_content)
graph.render('foodie_graph', format='png', cleanup=True)  # Save the graph as a PNG file
graph.view()  # View the graph in a new tab
```
